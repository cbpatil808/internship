### importing 

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import seaborn as sns
```

### Reading the data

```python
data = pd.read_csv(r"C:\Users\Chetan\Downloads\New_folder/terrorism.csv",encoding='ISO-8859-1')
data.head(10)
```

# Total number of terrorism activity per year

```python
data['iyear'].value_counts(dropna = False).sort_index()
```

# Visual Representation

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)

plt.plot(data['iyear'].unique(),data['iyear'].value_counts(dropna = False).sort_index(),marker='o')
a = list(data['iyear'].unique())

plt.xticks(a[::2])

plt.title("Global Terrorism",size=25,c='blue')
plt.xlabel("Year",size=20,c='red')
plt.ylabel("No of Attacks",size=20,c='red')

plt.show()
```

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)

sns.barplot(data['iyear'].unique(),data['iyear'].value_counts(dropna = False).sort_index())
plt.title("Global Terrorism",size=25,c='blue')
plt.xlabel("Year",size=20,c='red')
plt.ylabel("No of Attacks",size=20,c='red')
plt.show()

```

```python
fig=plt.subplots(figsize=(10,10))

plt.pie(data['iyear'].value_counts(dropna = False).sort_index(),labels=data['iyear'].unique())
plt.legend(loc='upper right',bbox_to_anchor=(0,1))
```

```python
pd.crosstab(data.iyear, data.region_txt).plot(figsize=(20,8))
plt.title('Terrorist activity by region',size=25,c='red')
plt.ylabel('Number of Attacks',size=20,c='green')
plt.xlabel("Year",size=20,c='green')

```

### Top-10 Terrorism Country

```python
print("Country       total number of terrorist activities")
data.country_txt.value_counts()[:10]
```

### Visual Representation

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)
sns.barplot(data['country_txt'].value_counts()[:10].index,data['country_txt'].value_counts()[:10].values)
plt.show()
```

```python
res = data[data['country_txt']=='India']
res.head(10)
```

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)
plt.plot(res['iyear'].unique(),res['iyear'].value_counts(dropna = False).sort_index(),marker='o')
plt.title("Terrorism activity in India",size=25,c='red')
plt.xlabel("Year",size=20,c='blue')
plt.ylabel("No of terrorist Activities",size=20,c='blue')
plt.show()
```
